<h4>Exploratory Data Analysis

```python
# Minimum Date
dataset['order_date'].min()
```

```python
# Maximum date
dataset['order_date'].max()
```

```python
# Getting month year from the dataset
dataset['month_year'] = dataset['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
dataset['month_year']
```

```python
df_month_year_trend = dataset.groupby('month_year').sum()['sales'].reset_index()
df_month_year_trend.head(50)
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_month_year_trend['month_year'], df_month_year_trend['sales'], color='Red')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

<h4>Which are the top 10 products by sale?

```python
dataset.head()
```

```python
dataset['product_name']
```

```python
# Grouping product name column by sales
dataset.groupby('product_name').sum()['sales']
```

```python
# Grouping the product by sales
product_sales = pd.DataFrame(dataset.groupby('product_name').sum()['sales'])
```

```python
# Sorting product name column
product_sales = product_sales.sort_values('sales', ascending=False)
```

```python
# top 10 products by sale
product_sales[:10]
```

<h4>Which are the most selling products?

```python
# Grouping the products by quantity
most_selling_product = pd.DataFrame(dataset.groupby('product_name').sum()['quantity'])
```

```python
most_selling_product.head()
```

```python
# Sorting the dataframe
most_selling_product = most_selling_product.sort_values('quantity', ascending=False)
most_selling_product.head()
```

```python
# Most selling products
most_selling_product[:10]
```

<h4> What is the most preferred ship mode ?

```python
# Setting the figure size
plt.figure(figsize=(10,8.5))

# Countplot: Counting the observations in each categorical bin using bars
sns.countplot('ship_mode', data=dataset)

# Display the figure
plt.show()
```

<h4> Which are the most profitable category and sub-category ?

```python
# Grouping by category and sub-category
category = pd.DataFrame(dataset.groupby(['category','sub_category']).sum()['profit'])
category
```

```python
# Sorting the values
category = category.sort_values(['category','profit'], ascending=False)
category
```
